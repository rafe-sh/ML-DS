## Matplotlib Object Oriented method

```python
### Create Figure
fig = plt.figure(figsize=(12,5), dpi=100)

x = np.arange(1,100)

### Creating Axis
ax1 = fig.add_axes([0,0,1,1])  # left, Bottom, Width, Height (ranges from 0 to 1)
ax2 = fig.add_axes([0.17, 0.2, 0.1, 0.2])

### Plot
ax1.plot(x, 1/13700*(x**2) + (1/x), '.g--')
ax2.plot(x, 1/13700*(x**2) + (1/x), '.b')

### Parameters
ax2.set_xlim(15,25)
ax2.set_ylim(0.075,0.085)
ax2.set_title('Zoomed IN', fontsize=10)

### Saving Figure
# fig.savefig('figure.png',bbox_inches='tight')  

plt.show()
```

## Matplotlib Subplots

`fig, axes = plt.subplots()`

```python
### Using `with` to inline style - doesnt affect other plots
with plt.style.context(['seaborn-darkgrid', 'ieee', 'no-latex']):
    
    fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

    ax[0][0].plot(x-50,x)
    ax[0][1].plot(x-50,(x-50)**2)
    ax[1][0].plot(x-50,(x-50)**3)
    ax[1][1].plot(x-50,(x-50)**4)

    ax[0][0].set_title('$x^1$')
    ax[0][1].set_title('$x^2$')
    ax[1][0].set_title('$x^3$')
    ax[1][1].set_title('$x^4$')
    
    # Manual Adjusting
    fig.subplots_adjust(left=None,
                        bottom=None,
                        right=None,
                        top=None,
                        wspace=0.15,
                        hspace=0.2,)

#     plt.tight_layout()
    plt.show()
```

## Matplotlib Styling

`lw` == Line width,  ranges from 0 to n  
`ls` == Line style,  options are `:`, `--`, `-.` , ...  
`marker` and `ms` == marker symbol and marker size,  
`alpha` == transparency,  ranges from 0 to 1  
`label` == the label of curvature, `plt.legend` is needed after that and the `loc` is its location argument  

```python
fig = plt.figure(figsize=(5,4), dpi=100)

plt.plot(x-50, (x-50)**4, label='$x^4$', ls='--', alpha=0.5, color='black', marker='+', ms=5)
plt.plot(x-50, (x-50)**2, label='$x^2$', ls='-.', alpha=0.6, color='#8B008B', ms=5, marker='.')

### Further adjustments
plt.grid(ls='--', color='blue')
plt.yscale('log')

plt.legend(loc=(1.05, 0.5))
plt.show()
```

### Gridspec

```python
import matplotlib.gridspec as gridspec

fig = plt.figure()

gs = gridspec.GridSpec(2, 3, height_ratios=[2,1], width_ratios=[1,2,1])
for g in gs:
    ax = fig.add_subplot(g)
    
fig.tight_layout()
```
