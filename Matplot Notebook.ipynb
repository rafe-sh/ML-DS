```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 2*np.pi,101)
```

```python
x = np.sin(t)
y = np.cos(t)
```

```python
plt.plot(x,y)

plt.title('Cicle')

plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.xlim(-1.5,1.5)
plt.ylim(-1.1,1.1)

plt.savefig('circle.jpg')

plt.show()
```

```python
# OOP PLOTTING
fig = plt.figure(figsize=(5,5),dpi=100)

axes1 = fig.add_axes([0,0,1,1])
axes1.set_title('Circle')
axes1.plot(x,y)

axes2 = fig.add_axes([1.1,0.5,0.5,0.5])
axes2.plot(x)

axes3 = fig.add_axes([1.1,0,0.5,0.5])
axes3.plot(y)

fig.savefig('Circle2.jpg', bbox_inches='tight')
plt.show()
```

```python
a = np.linspace(-2,10, 51)
```

```python
b = np.exp(a)
```

```python
fig , axes = plt.subplots(nrows=2, ncols=1, figsize= (8,5), dpi= 100)
fig.suptitle("Subplots", fontsize=15)

axes1 = fig.add_axes([0.15,0.15,0.2,0.15])
axes1.plot(a,b, 'b.')
axes1.set_xlim(-2,2)
axes1.set_ylim(0,2)
axes1.set_title('Zoomed In', fontsize=8)

axes[0].plot(x,y,'g.', label= 'circle', lw=2)
axes[0].legend(loc= 10)
axes[0].set_title('Circle')

axes[1].plot(a,b, 'r.-', label= 'a vs exp(a)')
axes[1].legend(loc= (1.1,0.4))
axes[1].set_title('Exponential')

plt.tight_layout()

fig.savefig('Subplots.png', bbox_inches= 'tight')

plt.show()
```

```python
fig, ax= plt.subplots()
ax.plot(a,b, 'b.-.', alpha=1)
ax.plot(a,b/2, color='red', alpha=0.5, ls= ':', marker= '+', ms=5)
```

```python
# Wrinting E = mc2
m = np.linspace(0,10,11)
c = 300000000
E = m * (c**2)
```

```python
plt.plot(m,E,'r',lw=5)
plt.title('$E=mc^2$')
plt.xlim(0,10)
plt.xlabel('mass (gr)')
plt.ylabel('enenrgy (J)')

plt.show()
```

```python
plt.plot(m, E,'r',lw=5)
plt.title('$E=mc^2$')
plt.xlabel('mass (gr)')
plt.ylabel('enenrgy (J)')

plt.yscale('log')   #### YSCALE

plt.grid(which='both',axis='y')
```

```python
labels = ['1 Mo','3 Mo','6 Mo','1 Yr','2 Yr','3 Yr','5 Yr','7 Yr','10 Yr','20 Yr','30 Yr']

july16_2007 =[4.75,4.98,5.08,5.01,4.89,4.89,4.95,4.99,5.05,5.21,5.14]
july16_2020 = [0.12,0.11,0.13,0.14,0.16,0.17,0.28,0.46,0.62,1.09,1.31]
```

```python
fig , ax= plt.subplots(figsize=(8,5))
ax.plot(labels, july16_2007, label='july16_2007')
ax.plot(july16_2020, label='july16_2020')
ax.legend(loc=(1.05,0.5));
```

```python
fig, ax= plt.subplots(nrows=2, ncols=1,figsize=(10,6))

ax[0].plot(labels, july16_2007)
ax[0].set_title('july16_2007')

ax[1].plot(labels, july16_2020)
ax[1].set_title('july16_2020')

plt.tight_layout()
```

```python
fig , ax= plt.subplots(figsize=(8,5),dpi=100)

### title
ax.set_title('July 16th Yiels Curves', fontsize=10)

###SPINES
ax.spines['left'].set_color('blue')
ax.spines['left'].set_linewidth(5) 

ax.spines['right'].set_color('red')
ax.spines['right'].set_linewidth(5)


ax.plot(labels, july16_2007,'b',lw=2)

axleft = ax.twinx()
axleft.plot(labels, july16_2020,'r',lw=2)

##LABELS
ax.set_ylabel('2007', color= 'blue')
axleft.set_ylabel('2020', color= 'red')

### Change yticklabels
for i in ax.get_yticklabels():
    i.set_color('b')
for i in axleft.get_yticklabels():
    i.set_color('r')
```
